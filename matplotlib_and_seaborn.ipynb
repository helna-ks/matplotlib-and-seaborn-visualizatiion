```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[5,6,3,4,7]
plt.scatter(x,y)

plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.plot(x,y,"*",linestyle="dotted",color="r",linewidth=5)
plt.show()
```

```python
subject=["maths","Science","malaylam","english","computer"]
marks=[34,56,12,78,36]
plt.bar(subject,marks,edgecolor="black",color=["green","red","yellow","blue","orange"])
plt.title("Bar Chart")
plt.xlabel("Subject",size=10)
plt.ylabel("Marks",size=10)
plt.show()
```

```python
subject=["maths","Science","malaylam","english","computer"]
marks=[34,56,12,78,36]
explode = [0.1, 0.2, 0.5, 0, 0]
colors = ( "orange", "cyan", "yellow","grey", "green",)
plt.pie(marks,labels=subject,explode=explode,autopct="%1.2f%%")
plt.title("Pie Chart")
plt.legend(loc="lower right",shadow=True)
plt.show()
```

```python
data=[10,23,45,10,23,67,56,23]
plt.figure(facecolor="yellow")
plt.hist(data,color="red")
plt.title("Histogram")

plt.xlabel("values")
plt.ylabel("Frequency")
plt.show()
```

```python
age=[23,56,21,5,50]
plt.boxplot(age)
plt.show()
```

```python
x=[1,2,4,5]
y=[4,8,9,2]
plt.subplot(1,2,2)
plt.plot(x,y)

x=[8,2,4,5]
y=[4,4,9,7]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.show()
```

```python
x=[1,5,3,4]
sales=[15,68,90,34]
plt.fill_between(x,sales)
plt.plot(x,sales)
plt.show()
```

```python
x=[1,2,4,5]
y=[4,8,9,2]
plt.stem(x,y)
```

```python
import matplotlib.pyplot as plt

data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
    [12,45,67]
]
plt.imshow(data, cmap='hot')
plt.colorbar()
plt.title("Simple Heatmap")
plt.show()
```

```python
data=sns.load_dataset('titanic')
```

```python
data
```

```python
data.head(3)
```

Countplot

```python
sns.countplot(x="sex",data=data,palette='pastel',hue="class")
plt.title("CountPlot graph")
plt.show()
```

Barplot

```python
sns.barplot(x="sex",y="fare",data=data)
plt.title("BarPlot graph")
plt.show()
```

Scatterplot

```python
sns.scatterplot(x="age",y="fare",data=data)
plt.title("ScatterPlot graph")
plt.show()
```

Lineplot

```python
sns.lineplot(x="age",y="sex",data=data)
plt.title("LinePlot graph")
plt.show()
```

```python
sns.boxplot(x="sex",y="age",data=data)
plt.title("BoxPlot graph")
plt.show()
```

```python
sns.histplot(x="age",data=data)
plt.title("HistPlot graph")
plt.show()
```

```python
sns.kdeplot(x="age",data=data)
plt.title("Kdeplot graph")
plt.show()
#
```
